```python
# Import pandas 
import pandas as pd

# Read in white wine data 
white = pd.read_csv("http://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-white.csv", sep=';')

# Read in red wine data 
red = pd.read_csv("http://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv", sep=';')

```

```python
# Print info on white wine
print(white.info())

# Print info on red wine
print(red.info())
```

```python
# First rows of `red` 
red.head()

# Last rows of `white`
white.tail()

# Take a sample of 5 rows of `red`
red.sample(5)

# Describe `white`
white.describe()

# Double check for null values in `red`
pd.isnull(red)

```

```python
#PREPARATION DES DONNEES
# Add `type` column to `red` with value 1
red['type'] = 1

# Add `type` column to `white` with value 0
white['type'] = 0

# Append `white` to `red`
wines = red.append(white, ignore_index=True)
```

```python
# Import `train_test_split` from `sklearn.model_selection`
from sklearn.model_selection import train_test_split

# Specify the data 
X=wines.iloc[:,0:11]

# Specify the target labels and flatten the array
y= np.ravel(wines.type)

# Split the data up in train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
# Import `StandardScaler` from `sklearn.preprocessing`
from sklearn.preprocessing import StandardScaler

# Define the scaler 
scaler = StandardScaler().fit(X_train)

# Scale the train set
X_train = scaler.transform(X_train)

# Scale the test set
X_test = scaler.transform(X_test)
```

```python
# Import `Sequential` from `keras.models`
from keras.models import Sequential

# Import `Dense` from `keras.layers`
from keras.layers import Dense

# Initialize the constructor
model = Sequential()


#Le réseau de neurones est constitué de 3 couches full connexion : une couche d'input, une couche cachée, et une couche de sortie.
#Relu : rectification linéaire
#Sigmoid : effet se seuil, réponse 0 ou 1

# Add an input layer 
model.add(Dense(12, activation='relu', input_shape=(11,)))

# Add one hidden layer 
model.add(Dense(8, activation='relu'))

# Add an output layer 
model.add(Dense(1, activation='sigmoid'))
```

```python
# Model output shape
model.output_shape

# Model summary
model.summary()

# Model config
model.get_config()

# List all weight tensors 
model.get_weights()
```

```python
#Compile and Fit

#Next, it’s time to compile your model and fit the model to the data: once again, make use of compile() and fit() to get this done.

#adam optimizer and the binary_crossentropy loss function. accuracy as output value to optimize.
#The optimizer and the loss are two arguments that are required if you want to compile the model.
#Some of the most popular optimization algorithms used are the Stochastic Gradient Descent (SGD), ADAM and RMSprop

model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
                   
model.fit(X_train, y_train,epochs=20, batch_size=1, verbose=1)
```

```python
#predict values

y_pred = model.predict(X_test)

print (y_pred)
```

```python
#evaluer le modèle

score = model.evaluate(X_test, y_test,verbose=1)

print(score)
```
